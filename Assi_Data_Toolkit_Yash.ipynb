```python
#Que 14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# Perform the following tasks using Matplotlib:
# a) Plot the 'Tem+perature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y-axis for 'Humidity').
# b) Label the x-axis as 'Date'.
# c) Set the title of the  plot as 'Temperature and Humidity Over Time

# Create a time-series dataset

date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
temperature = np.random.randint(20, 35, size=len(date_rng))
humidity = np.random.randint(50, 80, size=len(date_rng))

df = pd.DataFrame({'Date': date_rng, 'Temperature': temperature, 'Humidity': humidity})
df
```

```python
# Plot the data
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Set the title
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

```python
#Que 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
# tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.
# b) Overlay a line plot representing the normal distribution's probability density function (PDF).
# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
# d) Set the title of the plot as 'Histogram with PDF Ov

from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Overlay the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title
plt.title('Histogram with PDF Overlay')

# Display the plot
plt.show()
```

```python
#Que 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.?

#Ans. # Generate random arrays
x = np.random.rand(50) * 10 - 5  # Random numbers between -5 and 5
y = np.random.rand(50) * 10 - 5  # Random numbers between -5 and 5

# Determine quadrant for each point
quadrant = []
for i in range(len(x)):
  if x[i] >= 0 and y[i] >= 0:
    quadrant.append('Quadrant I')
  elif x[i] < 0 and y[i] >= 0:
    quadrant.append('Quadrant II')
  elif x[i] < 0 and y[i] < 0:
    quadrant.append('Quadrant III')
  else:
    quadrant.append('Quadrant IV')

# Create the scatter plot
sns.scatterplot(x=x, y=y, hue=quadrant)

# Add a legend
plt.legend()

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Display the plot
plt.show()
```

```python
#Que 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function

#Ans 18 .

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable output in notebook
output_notebook()

# Generate data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a figure
p = figure(title="Sine Wave Function", x_axis_label="x", y_axis_label="sin(x)")

# Add a line glyph
p.line(x, y, line_width=2)

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
#Que 19.  Using Bo+eh, generate a bar chart of randomly generated cateorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Cateorical
# Bar Chart

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool
import numpy as np

# Enable output in notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 21, size=len(categories))

# Create a figure
p = figure(title="Random Categorical Bar Chart", x_axis_label="Categories", y_axis_label="Values")

# Add bars with color mapping
colors = ['red', 'green', 'blue', 'orange', 'purple']
p.vbar(x=categories, top=values, width=0.5, color=colors)

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@x"), ("Value", "@top")])
p.add_tools(hover)

# Show the plot
show(p)
```
